# Machine Learning Proyect

The idea is to give the probability to win the next match according with:
1. Number of mathces won
2. Number of matches played
3. Average number of serves
4. Number of tournaments won

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Now, generating the dataset, the next arrays are going to be:
- x_1: represent the number of mathces won (input)
- x_2: represent the number of mathces played (input)
- x_3: represent the average number of serves (input)(a number between 1 and 2)
- x_4: number of tournaments won (input)
- y: the probability to win the match (output)

```python
# Generating the dataset
num_samples = 1000

x_1 = np.array(np.random.randint(0, 101, num_samples))
x_2 = np.array(np.random.randint([x for x in x_1], 101, num_samples))
x_3 = np.array(np.random.uniform(1.0, 2.0, num_samples))
x_4 = np.array(np.random.randint(0, [max(min(x, 50),1) for x in x_1], num_samples))

# Creating the results, array y
y = np.zeros(num_samples)

for i in range(num_samples):
    if x_1[i] > 75 and x_2[i] > 70:
        if x_3[i] > 1.7 and x_4[i] > 25:
            y[i] = np.random.uniform(0.9, 1.0)
        else:
            y[i] = np.random.uniform(0.8, 0.9)
    elif x_1[i] > 50 and x_2[i] > 50:
        if x_3[i] > 1.5 and x_4[i] > 20:
            y[i] = np.random.uniform(0.7, 0.8)
        else:
            y[i] = np.random.uniform(0.6, 0.7)
    elif x_1[i] > 25 and x_2[i] > 30:
        if x_3[i] > 1.3 and x_4[i] > 15:
            y[i] = np.random.uniform(0.5, 0.6)
        else:
            y[i] = np.random.uniform(0.4, 0.5)
    else:
        if x_3[i] > 1.1 and x_4[i] > 10:
            y[i] = np.random.uniform(0.25, 0.4)
        else:
            y[i] = np.random.uniform(0.1, 0.25)
        

"""
for i in range(20):
    print(f"Player {i}")
    print(f"Matches Won {x_1[i]}")
    print(f"Matches Played {x_2[i]}")
    print(f"Average number of serves {x_3[i]}")
    print(f"Number of tournaments won {x_4[i]}")
    print(f"Winning Probability {y[i]}")
    print()

"""
```

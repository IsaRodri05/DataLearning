# Project: Students giving up courses

## Problem definition
In a educational platform, a lot of students are giving up their courses before finishing, causing some losses to the platform. For that reason, it's essential to analyse students' academic behaviour, which lets us know factors that might be associated with leaving.

## Objective
- Analyse academic variables for the students
- Build a model that predict the probability of leaving a course 

## Research questions

1. What proportion of students abandon their courses?
2. Which variables are associated with course abandonment?
3. Do students who study fewer hours leave courese more frequently?
4. Could course abandonment be predicted abandonment using machine learning?

The dataset consist in 24 columns and 3.020 entries. There are columns with null values such as player_level, membership_type and cacelled_at. Also, the dataset has 12 categorical (object) variables, 9 numerical variables and 3 boolean variables. Below the text, there is a explanation of each variable and its classification created by the analyst; the classification help to understand if the variable is meaningful to the analysis or not by identifying whether the variable was created before the reservation happen or not:

| Name variable | Meaning | Classification |
|----|-----|-----|
| 1. **reservation_id:** | Unique identifier of each reservation. | Available before the event |
| 2. **customer_id:** | Unique identifier of each client. | Available before the event |
| 3. **club_city:** | categorical variable that shows the name's city of club | Available before the event |
| 4. **court_type:** | categorical variable that indicates the surface type of the court (Synthethic, Hard, Clay) | Available before the event |
| 5. **reservation_created_at:** | datetime variable that shows the date and time when the reservation was created | Available before the event |
| 6. **reservation_datetime:** | datetime variable that indicates the date and time chosen by the customer to use the court | Available before the event |
| 7. **match_type:** | categorical variable that shows the type match (Single or Doubles) | Available before the event |
| 8. **player_level:** | categorical variable that indicates the customer's level (Beginner, Intermediate, Advanced). It could be defined by the player or the coach | Available before the event | 
| 9. **membership_type:** | categorical variable that indicates the memebership type of customer. It could be NULL if the client doesn't have a membership | Available before the event |
| 10. **rain_probability:** | numerical variable that shows the probability of rain during the reservation | Available before the event |
| 11. **temperature_c:** | numerical variable that indicates the estimated temperature in °C for the reservation | Available before the event |
| 12. **coach_reserved:** | boolean variable that shows if the customer ask for a coach during the match | Available before the event |
| 13. **ball_machine:** | boolean variable that shows if the customer ask for a ball machine during the reservation | Available before the event |
| 14. **hours_reserved:** | numerical variable that indicates the duration (in hours) of the reservation | Available before the event |
| 15. **price_paid:** | numerical variable that shows the price of the reservation, the client must pay before the reservation. _It would be the variable to evaluate_ | Available before the event |
| 16. **coupon_used:** | boolean variable that indicates if the customer use a coupon for its reservation | Available before the event | 
| 17. **payment_method:** | categorical variable that shows the payment method that was used | Available before the event |
| 18. **reservation_status:** | categorical variable that indicates the status of the reservation | Available after the event |
| 19. **refund_amount:** | numercical variable that shows the refund amount to the customer | Available after the event |
| 20. **cancelled_at:** | date variable that indicates the date and time when customer cancelled the reservation. If the customer doesn't cancell the reservation, the entrie must be NULL | Available after the event |
| 21. **customer_rating:** | numerical variable that shows the rate which was given by the customer. It's a number which is between 1 to 5. It could be NULL if the customer doesn't answer a form | Available after the event |
| 22. **marketing_source:** | categorical variable that indicates the marketing source for whereby the customer knew about CanchaMatch | Available before the event |
| 23. **notes:** | categorical variable that shows some note per reservation. It was created by the members staff and it could be NULL | Available before or after the event |
| 24. **internal_batch_id:** | numerical variable that identifies the entrie in the software | Not applied |
